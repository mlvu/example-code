```python
RED  = '#C82506'
BLUE = '#0365C0'
GREEN = '#00882B'
ORANGE = '#DE6A10'
PURPLE = '#773F9B'
YELLOW = '#DCBD23'

import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def clean(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    
def ca(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off', top='off', right='off', left='off',
    labelbottom='off', labelleft='off') # labels along the bottom edge are off
    

def center(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    labelbottom='off', labelleft='off')
    
    ax.set_aspect('equal')
```

```python
from scipy.stats import norm

plt.figure(figsize=(6, 4.5))

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

x = np.linspace(-3, 1, 1000)
ax1.plot(x, norm.pdf(x, scale=0.5, loc=-1), lw=3, alpha=1, label='probability density', color=BLUE)
ax1.set_title('probability density')
ax1.spines['bottom'].set_position('zero')
clean(ax1)

ax2.plot(x, norm.cdf(x, scale=0.5, loc=-1), lw=3, alpha=1, label='probability density', color=BLUE)
ax2.set_title('cumulative density $P(X < x)$')
ax2.spines['bottom'].set_position('zero')
clean(ax2)

plt.savefig('cdf.pdf', transparent=True)
```

```python
from scipy.stats import norm

plt.figure(figsize=(6, 4.5))

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

x = np.linspace(-3, 1, 1000)
ax1.plot(x, norm.pdf(x, scale=0.5, loc=-1), lw=3, alpha=1, label='probability density', color=BLUE)
ax1.set_title('probability density')
ax1.spines['bottom'].set_position('zero')
clean(ax1)

ax2.plot(x, norm.cdf(x, scale=0.5, loc=-1), lw=3, alpha=0.5, label='probability density', color=BLUE)
ax2.set_title('cumulative density $P(X < x)$')
ax2.spines['bottom'].set_position('zero')
clean(ax2)

# plt.savefig('cdf.pdf')

n = 5
sample = norm.rvs(size=n, scale=0.5, loc=-1)
ax1.scatter(sample, np.zeros(n), color=GREEN, s=20)

levels = np.linspace(0, 1, n + 2)  # endpoint 1 is included by default
plt.step(sorted(list(sample) + [max(sample), 1]), levels, color=GREEN) 

plt.tight_layout()

plt.savefig('cdf.5.pdf')

n = 50
sample = norm.rvs(size=n, scale=0.5, loc=-1)
ax1.scatter(sample, np.zeros(n), color=RED, s=10)

levels = np.linspace(0, 1, n + 2)  # endpoint 1 is included by default
plt.step(sorted(list(sample) + [max(sample), 1]), levels, color=RED) 

plt.tight_layout()

plt.savefig('cdf.50.pdf')

n = 500
sample = norm.rvs(size=n, scale=0.5, loc=-1)
ax1.scatter(sample, np.zeros(n), color='k', s=1)

levels = np.linspace(0, 1, n + 2)  # endpoint 1 is included by default
plt.step(sorted(list(sample) + [max(sample), 1]), levels, color='k') 

plt.tight_layout()

plt.savefig('cdf.500.pdf')
```
