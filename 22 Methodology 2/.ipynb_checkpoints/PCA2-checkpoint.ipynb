```python
RED  = '#C82506'
BLUE = '#0365C0'
GREEN = '#00882B'
ORANGE = '#DE6A10'
PURPLE = '#773F9B'
YELLOW = '#DCBD23'

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

def clean(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    
def ca(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off', top='off', right='off', left='off',
    labelbottom='off', labelleft='off') # labels along the bottom edge are off
    
def center(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    labelbottom='off', labelleft='off')
    
    ax.set_aspect('equal')
    
    
```

```python
x = np.random.randn(200, 2) * np.asarray([[0.35, 1.5]]) + np.asarray([[1.35, -0.4]])

plt.figure(figsize=(8, 12))

plt.subplot(321)

plt.scatter(x[:, 0], x[:, 1], s=3, color=BLUE)
plt.xlim((-5, 5))
plt.ylim((-5, 5))

plt.title('before')

center()

x = x - x.mean(axis=0)
x = x/ x.std(axis=0)

plt.subplot(322)

plt.scatter(x[:, 0], x[:, 1], s=3, color=BLUE)
plt.xlim((-5, 5))
plt.ylim((-5, 5))

plt.title('after')

center()

plt.subplot(323)

x = np.random.randn(200, 2) * np.asarray([[1.5, 0.35]]) + np.asarray([[-2.2, 0.5]])

plt.scatter(x[:, 0], x[:, 1], s=3, color=GREEN)
plt.xlim((-5, 5))
plt.ylim((-5, 5))
center()

x = x - x.mean(axis=0)
x = x/ x.std(axis=0)

plt.subplot(324)

plt.scatter(x[:, 0], x[:, 1], s=3, color=GREEN)
plt.xlim((-5, 5))
plt.ylim((-5, 5))

center()

plt.subplot(325)

x = np.random.randn(200, 2) * np.asarray([[1.5, 0.35]])

x = x.dot(np.asarray([[1, 0.5],[0.5, 1]])) + np.asarray([[-2.5, 1.05]])

plt.scatter(x[:, 0], x[:, 1], s=3, color=RED)
plt.xlim((-5, 5))
plt.ylim((-5, 5))
center()

x = x - x.mean(axis=0)
x = x/ x.std(axis=0)

plt.subplot(326)

plt.scatter(x[:, 0], x[:, 1], s=3, color=RED)
plt.xlim((-5, 5))
plt.ylim((-5, 5))

center()

plt.savefig('whitening.pdf')
```

```python
x = np.random.randn(200, 2) 

plt.figure(figsize=(8, 12))

plt.subplot(321)

plt.scatter(x[:, 0], x[:, 1], s=3, color=RED)
ca()

plt.savefig('standard.pdf')

x = np.random.randn(200, 2) * np.asarray([[1.5, 0.35]])

plt.figure(figsize=(8, 12))

plt.subplot(321)

plt.scatter(x[:, 0], x[:, 1], s=3, color=RED)
center()
ca()

plt.savefig('stretched.pdf')
```

```python
mu = [1.0, -1.0]

x = np.random.multivariate_normal(mu, [[2, 0.7],[0.7, 0.7]], 200)

plt.figure(figsize=(8, 8))

plt.scatter(x[:, 0], x[:, 1], s=12, color=BLUE)
plt.scatter(mu[0], mu[1], s=50, marker='d', color=ORANGE)
plt.xlim((-5, 5))
plt.ylim((-5, 5))

center()

plt.savefig('mvn-example.pdf')
```
