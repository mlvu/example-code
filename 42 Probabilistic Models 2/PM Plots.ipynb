```python
RED  = '#C82506'
BLUE = '#0365C0'
GREEN = '#00882B'
ORANGE = '#DE6A10'
PURPLE = '#773F9B'
YELLOW = '#DCBD23'

import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def clean(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    
def ca(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off', top='off', right='off', left='off',
    labelbottom='off', labelleft='off') # labels along the bottom edge are off
    

def center(ax=None):
    if ax is None:
        ax = plt.gca()
    ax.spines['left'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    labelbottom='off', labelleft='off')
    
    ax.set_aspect('equal')
```

```python
x = np.linspace(-5, 5, 1000)


y2 = 1 / math.sqrt(2 * math.pi) * np.exp(-0.5*(x ** 2))

plt.figure(figsize=(12, 4.5))
plt.plot(x, y2, label='$y = e^{-x^2}$')

clean()

ax = plt.gca()
ax.spines['left'].set_position('center')
ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    labelbottom='off', labelleft='off')

plt.ylim([0, None])
    
plt.savefig('normal.pdf')
```

```python


x = np.linspace(0, 5, 1000)

y1 = np.exp(-x)
y2 = np.exp(-(x ** 2))

plt.figure(figsize=(6, 4.5))
plt.plot(x, y1, label='$y = e^{-x}$', color=BLUE)
plt.plot(x, y2, label='$y = e^{-x^2}$', color=ORANGE)

plt.ylim([0, 1.1])

plt.legend()
clean()

plt.savefig('construction.1.pdf')
```

```python
x = np.linspace(-3, 3, 1000)

y2 = np.exp(-(x ** 2))
der = np.exp(-(x ** 2)) * (- 2 * x)

plt.figure(figsize=(6, 4.5))
plt.plot(x, y2, label='$y = e^{-x^2}$', color=BLUE)
plt.plot(x, der, label='derivative', color=ORANGE)


plt.legend()
clean()

ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')


plt.ylim([-1.1, 1.1])


ax.tick_params(
    axis='both',       # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    labelleft='off')

plt.savefig('inflection-points.pdf')
```

```python
from numpy.core.umath_tests import inner1d

x1r = (-3, 3)
x2r = (-3, 3)

A = np.asarray([[0.5, 0.2],[0.2, 0.5]])
t = np.asarray([0.5, 0.5])
S = A.dot(A.T)
det = np.linalg.det(S)

h = 1000 # step nums in the mesh
x1, x2 = np.meshgrid(np.arange(x1r[0], x1r[1], (x1r[1] - x1r[0])/h),
                     np.arange(x2r[0], x2r[1], (x2r[1] - x2r[0])/h))

x = np.column_stack((x1.ravel(), x2.ravel()))

x = x - t

print(x.shape)

xp = x.dot(np.linalg.inv(S))


values = (1/math.sqrt((2*math.pi)**2 * det))*np.exp(-0.5 * inner1d(xp, x))

values = values.reshape(x1.shape)

# 3D plot for good measure

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(20, 80)

ax.plot_surface(x1, x2, values, rstride=50, cstride=50, cmap='copper')

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$y$')
                              
ax.set_zticks([])
ax.set_zlim([0, 1.1])

plt.savefig('mvn.construction.4.pdf')
```

```python
from scipy.stats import norm

w1, w2, w3 = 0.1, 0.6, 0.4

c1 = norm(loc=3, scale=0.75)
c2 = norm(loc=6, scale=0.5)
c3 = norm(loc=7.5, scale=1.2)

x = np.linspace(2, 10, 1000)

plt.figure(figsize=(6, 4.5))

plt.plot(x, w1*c1.pdf(x), color=RED)
plt.plot(x, w2*c2.pdf(x), color=GREEN)
plt.plot(x, w3*c3.pdf(x), color=BLUE)

clean()

plt.savefig('components.pdf')
```
